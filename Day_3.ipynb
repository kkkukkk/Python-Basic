```python
# 입력) 당신의 점수를 입력하세요 >>>
# 출력) 당신의 점수는 95 점이고 학점은 A0 학점입니다.
# 0점~3점 -
# 4점~6점 O
# 7점~9점 +


score = input('당신의 점수를 입력하세요 >>> ')

if int(score) >= 90:
  grd = 'A'
elif int(score) >= 80:
  grd = 'B'
elif int(score) >= 70:
  grd = 'C'
elif int(score) >= 60:
  grd = 'D'
else:
  grd = 'F'

if int(score) >= 60:
  if int(score[1]) >= 7:
    grd += '+'
  elif int(score[1]) >= 4:
    grd += 'O'
  else:
    grd += '-'

print(f'당신의 점수는 {score}이고 학점은 {grd}학점입니다.')
```

```python
#################
# While
#################

# 반복문은 작업 수행을 한 번이 아니라 계속해서 수행해야 할 때 사용한다.
# while 반복 작업을 몇 번 수행할지 확실하지 않을 때 주로 사용한다.
# ex) 운동장을 해가 질 때까지 반복해서 돌자

# while 조건식 :
#  실행구문

# while True:
#  print('운동장 돌기')
```

```python
# n = 1 1부터 10까지 1씩 증가시키면서 반복

n = 1
while n <= 10:
  print(n) # 조건식을 만족할때 출력이므로 10까지 출력
  n += 1 # n = n + 1 (자바 n++)

print('--------------')
print(n) # 11이 되어 조건식을 불만족 하므로 out된 n 값은 11
```

```python
# 사용자로부터 임의의 정수를 입력받아 모두 리스트에 보관합니다.
# 단, 0을 입력하변 프로그램을 종료합니다.

list=[]
num = 1

while num != 0:
  num = int(input('정수입력 >>> '))
  
  if num != 0:
    list.append(num)

print(list)
```

```python
# while 중첩
# while 내부에 또 다른 while을 사용하는 것을 중첩이라고 한다.

# 3일동안 1~3교시 수업을 한다.
# 1일차 1교시입니다.
# 1일차 2교시입니다.
# 1일차 3교시 입니다.
# ....
# 3일차 3교시

day = 1

while day <= 3: 
  cls = 1
  while cls <= 3:
    print(f'{day}일차 {cls}교시입니다.')
    cls += 1
  day += 1
```

```python
# 구구단을 2단부터 9단까지 출력하세요

# 2 X 1 = 2

n1 = 2
while n1 <= 9:
  n2 = 1
  while n2 <= 9:
    print(f'{n1}X{n2}={n1*n2}')
    n2 += 1
  n1 += 1
```

```python
# 커피 1잔을 300원에 판매하는 자판기가 있다.
# 자판기에 돈을 넣었을 때 자판기에서 뽑을 수 있는 커피가 몇 잔이고
# 잔돈이 얼마인지 함께 출력하세요

mon = int(input('자판기에 얼마를 넣을까요 >>> '))
cfe = 1

while mon >= cfe*300 :
  print(f'커피{cfe}잔, 잔돈{mon - 300 * cfe}원')
  cfe += 1
  mon - 300
```

```python
# 비밀 번호(str)를 입력받아서 숫자와 문자가 모두 반드시 하나 이상
# 포함된 경우만 '사용 가능한 비번입니다.'
# .isalpha() - 알파벳으로만 구성되면 True
# .isnumeric() - 숫자로만 구성되면 True

pw = input('비밀 번호를 입력하세요 >>> ')
if pw.isalpha() != True and pw.isnumeric() != True:
  print('사용 가능한 비번입니다.')
else:
  print('불가능한 비번입니다.')
```

```python
# for 문 사용
pw = input('비밀번호를 입력하세요 >>>')
ch_count = 0
no_count = 0

for ch in pw:
  if ch.isalpha():
    ch_count += 1
  elif ch.isnumeric():
    no_count += 1

if ch_count > 0 and no_count > 0:
  print('사용 가능한 비번입니다.')
else :
  print('사용 불가능한 비번입니다.')
```

```python
# 사용자로부터 양의 정수를 하나 입력받아서 1부터 입력 받은 정수까지의
# 모든 정수의 합계를 구하는 프로그램을 작성하세요

no = int(input('임의의 양수를 입력하세요 >>> '))

sum = 0

for i in range(1,no+1):
  sum += i

print(f'1부터 {no}사이의 모든 정수의 합계는 {sum}입니다.')
```

```python
# 사용자로부터 임의의 정수를 하나 입력 받아서 그 숫자 만큼의
# '과일이름'입력 받아서 'basket' 리스트에 저장하는 프로그램을 작성하세요

# 몇 개의 과일을 보관할까요 ? >>>
# 1번째 과일을 입력하세요
# n번쨰 과일을 입력하세요

basket = []
num = int(input('몇 개의 과일을 보관할까요? >>> '))

for i in range (1,num+1):
  fruit = input(f'{i}번째 과일을 입력하세요 >>> ')
  basket.append(fruit)

print(f'입력받은 과일들은 {basket} 입니다.')
```
